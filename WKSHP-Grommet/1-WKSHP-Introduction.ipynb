# LAB 1: Getting started with Grommet


## What is Grommet?

Grommet is an open source React component library by Hewlett Packard Enterprise (HPE) that provides responsive, accessible, and mobile-first UI components for web projects. Grommet provides powerful theming tools that let users tailor the component library to align with own color, type, component interaction, and layout needs.

Grommet offers a simple, consistent, and thoughtfully designed user interfaces (UI) that can seamlessly integrate into business applications and streamlining operations. HPE is a huge proponent of simplifying complex environments. That is why HPE developed Grommet, and made it available to anyone looking to create consistent user experiences for their applications.

Grommet is both liked and used, sustained by a vibrant, grassroots community. It has attracted plenty of developer input with over 5k commits and 8.3k stars from a wide variety of different contributors.

Visit the [Grommet](https://v2.grommet.io) website, and [Grommet GitHub](https://github.com/grommet/grommet) repository for more information.

## Grommet Storybook

The storybook allows developers to document and users to explore Grommet components. It enables users of Grommet components to explore each component in isolation and test them in different states. Most importantly, Grommet Storybook serves as living styles that help both designers and developers synchronize around the patterns and behavior of each Grommet component.

To see examples for each Grommet component, visit the [Grommet Storybook](https://storybook.grommet.io/) website for more information.

## Grommet Code Sandbox

CodeSandbox is an online editor that's built for web application development in the browser. It allows developers to focus on writing code instead of setting up their development environments.

For developers willing to try out Grommet in a fast and easy manner, visit the [Grommet Code Sandbox](https://codesandbox.io/u/grommetux/sandboxes) website for more information.

## Who's using Grommet?

Grommet is used by both internal HPE and external developers to build enterprise-class and consumer-grade applications that can be used in web, desktop, and mobile-friendly formats. It underpins many of the company’s software-defined platforms, like GreenLake Platform, HPE Synergy, HPE Storage, Aruba Networks, ChatHPE, HPE Server management, etc.
