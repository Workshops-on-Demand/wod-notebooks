## Lab 4: Deploy your App to Netlify

This Jupyter notebook will drive you through the following steps :

- Sign up for a Netlify account using GitHub account
- Link your GitHub repository to Netlify
- Edit deployment settings
- Deploy your application on the internet

### Netlify?
Netlify is simply used for hosting static sites. Another alternative to static site hosting is GitHub pages. 

- Netlify allows user to generate a custom URL for a project under Netlify.app domain name e.g `studentxxx.netlify.app`
- It is possible to connect own custom domain e.g `mywebsite.com`, set up a CNAME, and A-Records on Netlify. More information from [Netlify website](https://www.netlify.com/blog/2020/03/26/how-to-set-up-netlify-dns-custom-domains-cname-and-a-records/)
- It is possible to drag and drop a folder containing project HTML, CSS, JS into Netlify for manual deployment using Netlify drop (screenshot below).
- For automatic deployment, a user can connect a GitHub repository to Netlify and have it automatically trigger new deployments with new repository commits in a continuous integration approach.
