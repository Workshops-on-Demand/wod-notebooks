### Theme option

As you finished designing your app, let's see how to apply/change to a different theme from the `hpe` theme that we started with. 

By default, the grommet theme will be selected in the designer.grommet.io, and you will see all your buttons, links etc in violet color. To select a different theme, click on your design name "StudentXXX", click on Configure, a pop up window will show. Under theme, select `grommet` for example. You will notice that the overall app theme and color has changed from green to violet. Select `hpe` theme again and close the pop window.
