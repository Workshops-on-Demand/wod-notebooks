# Introduction to the HPE Ezmeral ML Ops
HPE Ezmeral Container Platform (HPE ECP) uses container technology to make it simpler and more cost-effective to deploy, run and manage both cloud-native and non-cloud-native applications for machine learning, deep learning, and big data analytics use cases.

HPE Ezmeral Machine Learning (ML) Ops extends the speed and agility of HPE ECP by bringing the power of containers to the entire ML lifecycle to enable users to build, train, share, deploy, and monitor ML models. Ezmeral ML Ops supports sandbox development (notebooks), distributed training, and the deployment and monitoring of trained models in production. To enable sharing model and data artifacts, Ezmeral ML Ops integrates Project repository, source control, and model registry capabilities.

You will get started with some fundamental knowledge about how to interact with the HPE ECP control plane UI. We walk you, a tenant/project user, through step by step details to build, train, register and deploy the model. At the end of the session, you will be able to test model serving.

# Lab Environment - HPE Ezmeral Container Platform Architecture

This high-level architecture diagram with its compute and storage components is shown below. 

![HPE Ezmeral Arch diagram](Pictures/HPEEzmeralArchdiagram.png)
      

Our HPE Ezmeral Controller Platform deployment includes:
* The `Controller host` manages all the hosts that comprise the HPE Ezmeral Container Platform deployment.
* The `EPIC hosts` are under the direct control of the Controller host, orchestrating the containerized applications.
* The `Gateway host` acts as a proxy server that carries client requests (HPE ECP UI, REST API, K8s API (kubectl commands)) to the HPE ECP controller, to EPIC hosts or one of the K8s clusters the containerized application services running in EPIC or K8s clusters. Containerized application service endpoints are exposed from HPE ECP to users via the gateway re-mapped ports. 
* The `HPE Ezmeral Data Fabric` (a MapR File System) is a storage provider for persistent volumes for the applications that require persistence of data. The default StorageClass is available out of the box from the HPE Ezmeral Data Fabric (MapR) using the HPE Ezmeral Container Storage Interface (CSI) driver for MapR.
